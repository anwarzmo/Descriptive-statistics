# Cleaning the dataset 

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(0)

# Number of rows present in the dataset
num_rows = 2000

# Generate synthetic data for the dataset
data = {
    'Age': np.random.randint(18, 65, size=num_rows),  # Random ages between 18 and 65
    'Height (cm)': np.random.normal(170, 10, size=num_rows),  # Random heights with mean 170cm and std deviation 10cm
    'Weight (kg)': np.random.normal(70, 15, size=num_rows),  # Random weights with mean 70kg and std deviation 15kg
    'Income (USD)': np.random.randint(20000, 100000, size=num_rows),  # Random income between 20000 and 100000 USD
    'Score': np.random.uniform(0, 100, size=num_rows)  # Random score between 0 and 100
}

# Creating a DataFrame with the data
df = pd.DataFrame(data)

# Creating a random percentage of missing values for height and weight to allow for data cleaning
# Making 5% of height and weight values missing
missing_pct = 0.05
num_missing = int(num_rows * missing_pct)

random_rows = np.random.choice(df.index, size=num_missing, replace=False)
df.loc[random_rows, 'Height (cm)'] = np.nan
df.loc[random_rows, 'Weight (kg)'] = np.nan

# Display the DataFrame with missing values
print("DataFrame with Missing Values:")
print(df.head(10))  # Display first 10 rows

# Data Cleaning: Removing The Missing Values
cleaned_df = df.dropna()  # Drop rows with any missing values (NaNs)

# Display the cleaned DataFrame
print("\nCleaned DataFrame (After Removing Missing Values):")
print(cleaned_df.head(10))  # Display first 10 rows

# Saving the cleaned DataFrame to a CSV file
cleaned_csv_file_path = 'cleaned_dataset.csv'
cleaned_df.to_csv(cleaned_csv_file_path, index=False)

print(f"\nCleaned dataset saved to '{cleaned_csv_file_path}'")
```


# Calculating Descriptive statistics: 


```python

# Calculate Descriptive Statistics for Each Column Separately
columns_to_analyze = ['Age', 'Height (cm)', 'Weight (kg)', 'Income (USD)', 'Score']

for column in columns_to_analyze:
    # Extract the column data
    column_data = df[column]
    
    # Calculate descriptive statistics
    mean = column_data.mean()  # Mean
    median = column_data.median()  # Median
    std_dev = column_data.std()  # Standard Deviation
    var = column_data.var()  # Variance
    q1 = column_data.quantile(0.25)  # 1st Quartile (25th percentile)
    q3 = column_data.quantile(0.75)  # 3rd Quartile (75th percentile)
    iqr = q3 - q1  # Interquartile Range (IQR)
    minimum = column_data.min()  # Minimum
    maximum = column_data.max()  # Maximum
    data_range = maximum - minimum  # Range
    
    # Display descriptive statistics for the current column
    print(f"\nStatistics for '{column}' column:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Variance: {var:.2f}")
    print(f"1st Quartile (Q1): {q1:.2f}")
    print(f"3rd Quartile (Q3): {q3:.2f}")
    print(f"Interquartile Range (IQR): {iqr:.2f}")
    print(f"Minimum: {minimum}")
    print(f"Maximum: {maximum}")
    print(f"Range: {data_range}")
```

# Box Plot for each individual column showing shape of distribution 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Number of rows in the dataset
num_rows = 2000

# Generate synthetic data for the dataset
data = {
    'Age': np.random.randint(18, 65, size=num_rows),  # Random ages between 18 and 65
    'Height (cm)': np.random.normal(170, 10, size=num_rows),  # Random heights with mean 170cm and std deviation 10cm
    'Weight (kg)': np.random.normal(70, 15, size=num_rows),  # Random weights with mean 70kg and std deviation 15kg
    'Income (USD)': np.random.randint(20000, 100000, size=num_rows),  # Random income between 20000 and 100000 USD
    'Score': np.random.uniform(0, 100, size=num_rows)  # Random score between 0 and 100
}

# Create DataFrame from the generated data
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Plot box plots for each column with annotations for mean, median, and IQR
plt.figure(figsize=(12, 8))  # Set the figure size

# Loop through each column and plot the box plot
for i, column in enumerate(df.columns, 1):
    plt.subplot(2, 3, i)  # Create subplot for each column
    plt.boxplot(df[column], vert=False)  # Plot box plot horizontally
    
    # Calculate mean, median, and IQR (Interquartile Range)
    mean_val = df[column].mean()
    median_val = df[column].median()
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    
    # Annotate the box plot with mean, median, and IQR
    plt.scatter(mean_val, 1, color='red', label='Mean', zorder=3)
    plt.scatter(median_val, 1, color='green', label='Median', zorder=3)
    plt.hlines(1, q1, q3, color='blue', label='IQR', zorder=2)
    
    # Add labels and title for each subplot
    plt.xlabel(column)
    plt.yticks([])  # Hide y-axis ticks
    plt.title(f'Box Plot of {column}')
    plt.legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

The box-plot diagrams show us the distribution of the middle 50% of al values represented in the rectangle box. Half the middle values lie in the interquartile ramge shown by the box. It shows us the minimum value, maximum value, IQR, mean, and median. 

# Skewness of the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(0)

# Number of rows in the dataset
num_rows = 2000

# Generate synthetic data for the dataset
data = {
    'Age': np.random.randint(18, 65, size=num_rows),  # Random ages between 18 and 65
    'Height (cm)': np.random.normal(170, 10, size=num_rows),  # Random heights with mean 170cm and std deviation 10cm
    'Weight (kg)': np.random.normal(70, 15, size=num_rows),  # Random weights with mean 70kg and std deviation 15kg
    'Income (USD)': np.random.randint(20000, 100000, size=num_rows),  # Random income between 20000 and 100000 USD
    'Score': np.random.uniform(0, 100, size=num_rows)  # Random score between 0 and 100
}

# Create DataFrame from the generated data
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Calculate skewness for each column
skewness_data = df.skew()

# Plot skewness and distribution for each column
for column in df.columns:
    plt.figure(figsize=(10, 6))  # Set the figure size
    
    # Plot histogram with seaborn
    sns.histplot(df[column], kde=True, color='skyblue', edgecolor='black')
    
    # Calculate mean, median, and IQR
    mean = df[column].mean()
    median = df[column].median()
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    
    # Plot mean, median, and IQR on the histogram
    plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label='Mean')
    plt.axvline(median, color='green', linestyle='dashed', linewidth=1.5, label='Median')
    plt.axvline(q1, color='purple', linestyle='dashed', linewidth=1.5, label='Q1')
    plt.axvline(q3, color='purple', linestyle='dashed', linewidth=1.5, label='Q3')
    plt.axvspan(q1, q3, color='gray', alpha=0.3, label='IQR')
    
    # Add labels and title
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column} (Skewness: {skewness_data[column]:.2f})')
    
    # Add legend
    plt.legend()
    
    # Show plot
    plt.show()
```

Skewness is a measure of the lack of symmetry in the distribution of data.Normal distribution is when the distribution curve is symmetrical and there is no skewness on either side. A distribution is said to be positively skewed if skewness is greater than 0. This means that the tail end of the values point more to the higher values. A distributuion is said to be negatively skewed if if the tail end of the values point more to the lower values. 
It is clear to see that all distribution curves are mostly evenly distributed. They are near perfect symmetry, with some being left skewed and some being right skewed. All skewness values are close to 0.

# Conclusions for data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Number of rows in the dataset
num_rows = 2000

# Generate synthetic data for the dataset
data = {
    'Age': np.random.randint(18, 65, size=num_rows),  # Random ages between 18 and 65
    'Height (cm)': np.random.normal(170, 10, size=num_rows),  # Random heights with mean 170cm and std deviation 10cm
    'Weight (kg)': np.random.normal(70, 15, size=num_rows),  # Random weights with mean 70kg and std deviation 15kg
    'Income (USD)': np.random.randint(20000, 100000, size=num_rows),  # Random income between 20000 and 100000 USD
    'Score': np.random.uniform(0, 100, size=num_rows)  # Random score between 0 and 100
}

# Create DataFrame from the generated data
df = pd.DataFrame(data)

# Compute correlation matrix
correlation_matrix = df.corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

Positive correlations indicate that as one variable increases, so too does the other. 
Negative correlations indicate that as one variable increases, the other variable decreases. 
From the graph shown above it is clear to she how the variables correlate to one another.
Age has a positive correlation with heught, weight, and income. But it has a negative correlation with score. As age increases, score seems to decrease.
